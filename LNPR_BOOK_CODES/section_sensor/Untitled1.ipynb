```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal   ###multivariatenormal###

irlidar = multivariate_normal(mean=np.array([19.86,729.3]).T, cov=np.array([[42, -10],[-10, 17.7]]))
```

```python
import numpy as np                ###contour###
 
x, y = np.mgrid[0:40, 710:750]     #2XY
pos = np.empty(x.shape + (2,))     #x40x402340x40x2
pos[:, :, 0] = x                                   #3x,y
pos[:, :, 1] = y
cont = plt.contour(x, y, irlidar.pdf(pos)) #x, y
cont.clabel(fmt='%1.1e')                         #

plt.show()   #
```
